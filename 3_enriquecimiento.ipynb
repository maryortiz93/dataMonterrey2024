```python
%matplotlib inline
import pandas as pd
```

```python
compras_df=pd.read_csv('data/clean_compras_2024.csv')
```

```python
# Expandimos la columna FECHA
```

```python
# Convertimos la fecha de Object a Datetime
compras_df['FECHA']=pd.to_datetime(compras_df['FECHA'])
```

```python
# Creamos una columna llamada día
compras_df['DIA']=compras_df['FECHA'].apply(lambda x:x.day)
# Creamos una columna llamada mes
compras_df['MES']=compras_df['FECHA'].apply(lambda x:x.month)
# Creamos una columna llamada Año
compras_df['ANIO']=compras_df['FECHA'].apply(lambda x:x.year)
compras_df.head()
```

### **PROYECTAR UNA VARIABLE CON OTRA**

```python
# Contamos la cantidad de veces que aparece cada beneficiario
beneficiario_sales=compras_df['BENEFICIARIO'].value_counts()
```

```python
# convertimos esos datos en una estructura
beneficiario_sales=compras_df['BENEFICIARIO'].value_counts().to_dict()
# Visualizamos los primeros 5 valores de la lista
list(beneficiario_sales.keys())[0:5]
```

```python
# Ventas por cada beneficiario
compras_df['BENEFICIARIO_SALES']=compras_df['BENEFICIARIO'].apply(lambda x : beneficiario_sales[x])
```

```python
compras_df.head()
```

```python
# Calculamos el valor promedio de gasto por categoria
sheet_mean_compras=compras_df.groupby('SHEET')['IMPORTE'].mean()
```

```python
sheet_mean_compras
```

```python
# Convertimos a un diccionario
sheet_mean_compras.to_dict()
```

```python
compras_df['IMPORTE_GASTO_PROMEDIO']=compras_df['SHEET'].apply(lambda x : sheet_mean_compras[x])
compras_df.head(4)
```

```python
# Calculamos el valor de importe por beneficiario
beneficiario_importe=compras_df.groupby('BENEFICIARIO')['IMPORTE'].sum()
beneficiario_importe.head(4)
```

```python
# Convertimos a un diccionario
beneficiario_importe.to_dict()
```

```python
compras_df['IMPORTE_BENEFICIARIO']=compras_df['BENEFICIARIO'].apply(lambda x : beneficiario_importe[x])
compras_df.head(4)
```

```python
compras_df.to_csv('data/compras_df_enriched.csv', index=None)
```
