```python
# Extraemos los datos relacionados a los contratistas desde el mes de Enero hasta Agosto del año 2024
df_contratistas_enero=pd.read_excel('01-INTERNET-ENE-2024.xlsx',sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas_enero['SHEET']='CONTRATISTAS'
df_contratistas_febrero=pd.read_excel('02-INTERNET-FEB-2024.xlsx',sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas_febrero['SHEET']='CONTRATISTAS'
df_contratistas_marzo=pd.read_excel('03-INTERNET-MZO-2024.xlsx',sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas_marzo['SHEET']='CONTRATISTAS'
df_contratistas_abril=pd.read_excel('04-INTERNET-ABR-2024.xlsx',sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas_abril['SHEET']='CONTRATISTAS'
df_contratistas_mayo=pd.read_excel('05-INTERNET-MAY-2024.xlsx',sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas_mayo['SHEET']='CONTRATISTAS'
df_contratistas_junio=pd.read_excel('06-INTERNET-JUN-2024.xlsx',sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas_junio['SHEET']='CONTRATISTAS'
df_contratistas_julio=pd.read_excel('07-INTERNET-JUL-2024.xlsx',sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas_julio['SHEET']='CONTRATISTAS'
df_contratistas_agosto=pd.read_excel('08-INTERNET-AGO-2024.xlsx',sheet_name='OBRA PUBLICA Y GV FDOS. ESP', skiprows=5)
df_contratistas_agosto['SHEET']='CONTRATISTAS'
```

```python
df_contratistas=pd.concat([df_contratistas_enero, df_contratistas_febrero, df_contratistas_marzo, df_contratistas_abril, df_contratistas_mayo, df_contratistas_junio, df_contratistas_julio, df_contratistas_agosto])
df_contratistas.head(5)
```

```python
# Extraemos los datos relacionados a los Gastos Varios desde el mes de Enero hasta Agosto del año 2024
# El mes de Abril tiene una O adicional en el nombre de la hoja
df_gastos_enero=pd.read_excel('01-INTERNET-ENE-2024.xlsx',sheet_name='GTS VARIOS', skiprows=5)
df_gastos_enero['SHEET']='GASTOS VARIOS'
df_gastos_febrero=pd.read_excel('02-INTERNET-FEB-2024.xlsx',sheet_name='GTS VARIOS', skiprows=5)
df_gastos_febrero['SHEET']='GASTOS VARIOS'
df_gastos_marzo=pd.read_excel('03-INTERNET-MZO-2024.xlsx',sheet_name='GTS VARIOS', skiprows=5)
df_gastos_marzo['SHEET']='GASTOS VARIOS'
df_gastos_abril=pd.read_excel('04-INTERNET-ABR-2024.xlsx',sheet_name='GTOS VARIOS', skiprows=5)
df_gastos_abril['SHEET']='GASTOS VARIOS'
df_gastos_mayo=pd.read_excel('05-INTERNET-MAY-2024.xlsx',sheet_name='GTS VARIOS', skiprows=5)
df_gastos_mayo['SHEET']='GASTOS VARIOS'
df_gastos_junio=pd.read_excel('06-INTERNET-JUN-2024.xlsx',sheet_name='GTS VARIOS', skiprows=5)
df_gastos_junio['SHEET']='GASTOS VARIOS'
df_gastos_julio=pd.read_excel('07-INTERNET-JUL-2024.xlsx',sheet_name='GTS VARIOS', skiprows=5)
df_gastos_julio['SHEET']='GASTOS VARIOS'
df_gastos_agosto=pd.read_excel('08-INTERNET-AGO-2024.xlsx',sheet_name='GTS VARIOS', skiprows=5)
df_gastos_agosto['SHEET']='GASTOS VARIOS'
```

```python
df_gastos_varios=pd.concat([df_gastos_enero, df_gastos_febrero, df_gastos_marzo, df_gastos_abril, df_gastos_mayo, df_gastos_junio, df_gastos_julio, df_gastos_agosto])
df_gastos_varios.head(5)
```

```python
# Extraemos los datos relacionados a los Servicios Profesionales desde el mes de Enero hasta Agosto del año 2024
df_servicios_enero=pd.read_excel('01-INTERNET-ENE-2024.xlsx',sheet_name='SERV PROF', skiprows=5)
df_servicios_enero['SHEET']='SERVICIOS'
df_servicios_febrero=pd.read_excel('02-INTERNET-FEB-2024.xlsx',sheet_name='SERV PROF', skiprows=5)
df_servicios_febrero['SHEET']='SERVICIOS'
df_servicios_marzo=pd.read_excel('03-INTERNET-MZO-2024.xlsx',sheet_name='SERV PROF', skiprows=5)
df_servicios_marzo['SHEET']='SERVICIOS'
df_servicios_abril=pd.read_excel('04-INTERNET-ABR-2024.xlsx',sheet_name='SERV PROF', skiprows=5)
df_servicios_abril['SHEET']='SERVICIOS'
df_servicios_mayo=pd.read_excel('05-INTERNET-MAY-2024.xlsx',sheet_name='SERV PROF', skiprows=5)
df_servicios_mayo['SHEET']='SERVICIOS'
df_servicios_junio=pd.read_excel('06-INTERNET-JUN-2024.xlsx',sheet_name='SERV PROF', skiprows=5)
df_servicios_junio['SHEET']='SERVICIOS'
df_servicios_julio=pd.read_excel('07-INTERNET-JUL-2024.xlsx',sheet_name='SERV PROF', skiprows=5)
df_servicios_julio['SHEET']='SERVICIOS'
df_servicios_agosto=pd.read_excel('08-INTERNET-AGO-2024.xlsx',sheet_name='SERV PROF', skiprows=5)
df_servicios_agosto['SHEET']='SERVICIOS'
```

```python
df_servicios=pd.concat([df_servicios_enero, df_servicios_febrero, df_servicios_marzo, df_servicios_abril, df_servicios_mayo, df_servicios_junio, df_servicios_julio, df_servicios_agosto])
df_servicios.head(5)
```

```python
# Extraemos los datos relacionados a Comunicacion desde el mes de Febrero hasta Agosto del año 2024
# El mes de enero no tiene la hoja de COMUNICACION, Abril, Mayo, Junio y Agosto tiene el nombre COMUNIC
df_comunicacion_febrero=pd.read_excel('02-INTERNET-FEB-2024.xlsx',sheet_name='COMUNICACION', skiprows=5)
df_comunicacion_febrero['SHEET']='COMUNICACION'
df_comunicacion_marzo=pd.read_excel('03-INTERNET-MZO-2024.xlsx',sheet_name='COMUNICACION', skiprows=5)
df_comunicacion_marzo['SHEET']='COMUNICACION'
df_comunicacion_abril=pd.read_excel('04-INTERNET-ABR-2024.xlsx',sheet_name='COMUNIC', skiprows=5)
df_comunicacion_abril['SHEET']='COMUNICACION'
df_comunicacion_mayo=pd.read_excel('05-INTERNET-MAY-2024.xlsx',sheet_name='COMUNIC', skiprows=5)
df_comunicacion_mayo['SHEET']='COMUNICACION'
df_comunicacion_junio=pd.read_excel('06-INTERNET-JUN-2024.xlsx',sheet_name='COMUNIC', skiprows=5)
df_comunicacion_junio['SHEET']='COMUNICACION'
df_comunicacion_julio=pd.read_excel('07-INTERNET-JUL-2024.xlsx',sheet_name='COMUNICACION', skiprows=5)
df_comunicacion_julio['SHEET']='COMUNICACION'
df_comunicacion_agosto=pd.read_excel('08-INTERNET-AGO-2024.xlsx',sheet_name='COMUNIC', skiprows=5)
df_comunicacion_agosto['SHEET']='COMUNICACION'
```

```python
df_comunicacion=pd.concat([df_comunicacion_febrero, df_comunicacion_marzo, df_comunicacion_abril, df_comunicacion_mayo, df_comunicacion_junio, df_comunicacion_julio, df_comunicacion_agosto])
df_comunicacion.head(5)
```

```python
# Extraemos los datos relacionados a Gastos de Representación desde el mes de Enero hasta Julio del año 2024
# Abril tiene una O adicional en el nombre de la hoja, Junio y Agosto no tiene datos de Gastos de Representacion
df_representacion_enero=pd.read_excel('01-INTERNET-ENE-2024.xlsx',sheet_name='GTS REPRESENT', skiprows=5)
df_representacion_enero['SHEET']='REPRESENTACION'
df_representacion_febrero=pd.read_excel('02-INTERNET-FEB-2024.xlsx',sheet_name='GTS REPRESENT', skiprows=5)
df_representacion_febrero['SHEET']='REPRESENTACION'
df_representacion_marzo=pd.read_excel('03-INTERNET-MZO-2024.xlsx',sheet_name='GTS REPRESENT', skiprows=5)
df_representacion_marzo['SHEET']='REPRESENTACION'
df_representacion_abril=pd.read_excel('04-INTERNET-ABR-2024.xlsx',sheet_name='GTOS REPRESENT', skiprows=5)
df_representacion_abril['SHEET']='REPRESENTACION'
df_representacion_mayo=pd.read_excel('05-INTERNET-MAY-2024.xlsx',sheet_name='GTS REPRESENT', skiprows=5)
df_representacion_mayo['SHEET']='REPRESENTACION'
df_representacion_julio=pd.read_excel('07-INTERNET-JUL-2024.xlsx',sheet_name='GTS REPRESENT', skiprows=5)
df_representacion_julio['SHEET']='REPRESENTACION'
```

```python
df_representacion=pd.concat([df_representacion_enero, df_representacion_febrero, df_representacion_marzo, df_representacion_abril, df_representacion_mayo, df_representacion_julio])
df_representacion.head(5)
```

```python
# Extraemos los datos relacionados a Servicios Personales desde el mes de Enero hasta Agosto del año 2024
df_personales_enero=pd.read_excel('01-INTERNET-ENE-2024.xlsx',sheet_name='SERV PERS', skiprows=5)
df_personales_enero['SHEET']='PERSONALES'
df_personales_febrero=pd.read_excel('02-INTERNET-FEB-2024.xlsx',sheet_name='SERV PERS', skiprows=5)
df_personales_febrero['SHEET']='PERSONALES'
df_personales_marzo=pd.read_excel('03-INTERNET-MZO-2024.xlsx',sheet_name='SERV PERS', skiprows=5)
df_personales_marzo['SHEET']='PERSONALES'
df_personales_abril=pd.read_excel('04-INTERNET-ABR-2024.xlsx',sheet_name='SERV PERS', skiprows=5)
df_personales_abril['SHEET']='PERSONALES'
df_personales_mayo=pd.read_excel('05-INTERNET-MAY-2024.xlsx',sheet_name='SERV PERS', skiprows=5)
df_personales_mayo['SHEET']='PERSONALES'
df_personales_junio=pd.read_excel('06-INTERNET-JUN-2024.xlsx',sheet_name='SERV PERS', skiprows=5)
df_personales_junio['SHEET']='PERSONALES'
df_personales_julio=pd.read_excel('07-INTERNET-JUL-2024.xlsx',sheet_name='SERV PERS', skiprows=5)
df_personales_julio['SHEET']='PERSONALES'
df_personales_agosto=pd.read_excel('08-INTERNET-AGO-2024.xlsx',sheet_name='SERV PERS', skiprows=5)
df_personales_agosto['SHEET']='PERSONALES'
```

```python
df_personales=pd.concat([df_personales_enero, df_personales_febrero, df_personales_marzo, df_personales_abril, df_personales_mayo, df_personales_junio,df_personales_julio,df_personales_agosto])
df_personales.head(5)
```

```python
# Unimos todos los datos en un solo dataframe llamado df_compras_2024
df_compras_2024=pd.concat([df_contratistas, df_gastos_varios, df_servicios, df_comunicacion, df_representacion, df_personales])
```

```python
# Visualizamos las 5 primeras filas
df_compras_2024.head(5)
```

```python
# Eliminamos los valores faltantes o nulos
df_compras_2024_com=df_compras_2024.dropna(how='any')
```
